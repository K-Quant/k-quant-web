```python
import pandas as pd
```

```python
df = pd.read_csv("temp.csv").drop(columns=['Unnamed: 0']).rename(columns={'label':'rela'})
```

```python
df.insert(df.shape[1], 'type', "resolved")
```

```python
new_list = []
```

```python
for i in range(len(df)):
    tmp_dict = {}
    tmp_dict["source"] = df["source"][i]
    tmp_dict["target"] = df["target"][i]
    tmp_dict["rela"] = df["rela"][i]
    tmp_dict["type"] = df["type"][i]
    # print(tmp_dict)
    new_list.append(tmp_dict)
```

```python
new_list
```

```python
import json
b = json.dumps(new_list,ensure_ascii=False)
f2 = open('graph.json', 'w')
f2.write(b)
f2.close()
```
